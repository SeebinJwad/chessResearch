```python
ab_pairs = []
ci = .02
for b in range(100):
    for a in range(b):
        # assuming the values must be within confidence interval of .7 and a, b odds (from looking at examples in prob)  
        if abs((a / b) - .7) < ci and a % 2 == 1 and b % 2 == 1:
            ab_pairs.append([a,b])
```

```python
# converting equation m(m+1)/8 = n(n+1) to pells equation
# common denom - inp = multiple, num = 64
def cd(inp, num=64):
    if inp % 64 != 0:
        return False, False
    return inp // 64, 64
multiple = 2048
m_outer = cd(multiple // 8)[0]
n_outer = cd(multiple)[0]
print(str(m_outer) + "(64m(m+1) = " + str(n_outer) + "(64n(n+1))")

complete_square_add_val = 4**2 * n_outer - 4**2 * m_outer
print(str(complete_square_add_val) + " + " + str(m_outer) + "((8m+4)^2) = " + str(n_outer) + "((8n+4)^2)")
```
